```python
#To show the percentage of overall values in the dataset and perentage of ones who have charged off of the entire Charged Off values
def showPercentage(filterVariable , colour=''):
    plt.figure(1 , figsize=(20,7))                
    plt.subplot(1, 2, 1)
    if colour:
        ax  = (loan[filterVariable].value_counts(normalize=True)*100).sort_index().plot.bar(color=colour)
    else:
        ax  = (loan[filterVariable].value_counts(normalize=True)*100).sort_index().plot.bar()
    ax.set(xlabel=filterVariable, ylabel='Total '+filterVariable+' in %')
    addLabels(ax)
    ax.title.set_text('Overall Distribution')
    plt.subplot(1, 2, 2)
    if colour:
        ax = (loan[loan['loan_status']=='Charged Off'][filterVariable].value_counts(normalize=True)*100).sort_index().plot.bar(color=colour)
    else:
        ax = (loan[loan['loan_status']=='Charged Off'][filterVariable].value_counts(normalize=True)*100).sort_index().plot.bar()
    ax.set(xlabel=filterVariable, ylabel='ChargedOff '+ filterVariable +' in %')
    addLabels(ax)
    ax.title.set_text('Distribution of Charged Off Loans')
    
    plt.tight_layout()
```

```python
# For each bar: Place a label
def addLabels(ax):
        for p in ax.patches:
            ax.annotate(format(round(p.get_height() , 2))+"%", 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha = 'center', va = 'center', 
                       size=12,
                       xytext = (0, 9), 
                       textcoords = 'offset points')
        return ax  
```

```python
def describeColumn(col):
    return loan.groupby(['loan_status'])[col].describe()
```

```python
# Show the percentage of charged off including both fully paid and charged off
def showPercentForColumn(indexCol):
    plt.figure(1 , figsize=(15,10))
    res1=pd.pivot_table(loan,columns="loan_status",index=indexCol, values='row_count',aggfunc=sum)
    res1["bin_total"]=res1["Charged Off"]+res1["Fully Paid"]
    res1['% Charged Off']=(res1['Charged Off']/res1["bin_total"])*100
    res1['% Charged Off'] = round(res1['% Charged Off'] , 2)
    ax = res1['% Charged Off'].plot.bar()
    ax.set(xlabel=indexCol, ylabel='Percent for column')
    addLabels(ax)
    ax.title.set_text("Contribution of Charge Off by {}".format(indexCol))
    plt.show()
    return ax
```
